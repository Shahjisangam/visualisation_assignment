Generate a Synthetic Dataset

We'll create a dataset with several correlated features using NumPy and Pandas. Let's assume we want to create a dataset with three features that are linearly correlated.

python

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
x = np.random.normal(0, 1, n_samples)
y = 2 * x + np.random.normal(0, 0.5, n_samples)  # y is correlated with x
z = -0.5 * x + 0.5 * y + np.random.normal(0, 0.5, n_samples)  # z is correlated with x and y

# Create a DataFrame
data = {
    'Feature1': x,
    'Feature2': y,
    'Feature3': z
}

df = pd.DataFrame(data)
```

 Visualize the Correlation Matrix Using a Heatmap

We'll use Seaborn to create a heatmap of the correlation matrix.

python

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = df.corr()

# Set the size of the plot
plt.figure(figsize=(8, 6))

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

# Add titles and labels
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

Explanation

    Synthetic Data: We generated three features (Feature1, Feature2, Feature3) with different levels of correlation.
    Correlation Matrix: This matrix shows the Pearson correlation coefficients between each pair of features.
    Heatmap: The heatmap visualizes the correlation matrix, with colors indicating the strength and direction of correlations. The annot=True parameter adds numerical values to each cell in the heatmap for clarity.

# PLOTLY ASSIGNMENT

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(30)

# Generate synthetic data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)
```

Create a 3D Scatter Plot

We'll use Matplotlib to create a 3D scatter plot.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set up the figure and 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot data points
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of X, Y, and Z')

# Show the plot
plt.show()
```

Explanation

    fig.add_subplot(111, projection='3d'): Creates a 3D subplot.
    ax.scatter: Plots the data points in 3D space.
    Labels and Title: Adds axis labels and a title for clarity.

Violin Plot for Student Grades

A violin plot is useful for visualizing the distribution of scores across different grade categories.
Code to Create the Violin Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data for student grades
np.random.seed(10)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data)

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df_grades, inner='quartile')

# Add titles and labels
plt.title('Distribution of Scores Across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')

# Show the plot
plt.show()
```

Heatmap for Sales Data

To visualize the variation in sales across different months and days, you can create a heatmap.
Code to Create the Heatmap

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data for sales
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

# Pivot the DataFrame for heatmap
sales_pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Create a heatmap
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5)

# Add titles and labels
plt.title('Sales Data Heatmap')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

Explanation

    Violin Plot: Shows the distribution of scores across different grade categories. The inner='quartile' parameter adds quartile information inside the violin plot.
    Heatmap: Visualizes the average sales across different days and months. The pivot_table function is used to transform the data into a format suitable for the heatmap, and cmap='YlGnBu' specifies the color map.

Step-by-Step Guide

    Generate the Sales Data

Let's start with generating the synthetic sales data:

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(20)

# Generate synthetic sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)
```

    Pivot the Data

We need to pivot the DataFrame to create a matrix suitable for the heatmap:

```python
# Pivot the DataFrame
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')
```

    Create the Heatmap

Use Seaborn to create a heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Create a heatmap
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5, cbar_kws={'label': 'Average Sales'})

# Add titles and labels
plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

Explanation

    Pivot Table: The pivot_table function reshapes the data such that Day values become rows, Month values become columns, and Sales values populate the cells.
    Heatmap: The sns.heatmap function visualizes the pivoted data. The cmap='YlGnBu' parameter sets the color scheme, annot=True adds numerical values to the cells, and fmt='.1f' formats the numbers with one decimal place. linewidths=0.5 adds lines between cells, and cbar_kws customizes the color bar.
